# Bước 1: Import các thư viện và khai báo các hằng số

## 1.1 Import

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## 1.2 Constant

```python
CN_INC = 'Income'  # income column name
CN_AGE = 'Age'  # age column name
CN_EXP = 'Experience'  # experience column name
CN_MS = 'Married/Single'  # married or single column name
CN_HO = 'House Ownership'  # house ownership column name
CN_CO = 'Car Ownership'  # car ownership column name
CN_PRO = 'Profession'  # profession column name
CN_ST = 'State'  # state column name
CN_RF = 'Risk Flag'  # risk flag column name
```

# Bước 2: Đọc dữ liệu và thống kê mô tả

## 2.1 Đọc dữ liệu

```python
df = pd.read_csv('data_valid.csv')

income = df[CN_INC]
age = df[CN_AGE]
experience = df[CN_EXP]
marr_sngl = df[CN_MS]
hse_or = df[CN_HO]
car_or = df[CN_CO]
profession = df[CN_PRO]
state = df[CN_ST]
risk_flag = df[CN_RF]

inc_vct = income.value_counts()
age_vct = age.value_counts()
exp_vct = experience.value_counts()
ms_vct = marr_sngl.value_counts()
ho_vct = hse_or.value_counts()
co_vct = car_or.value_counts()
pro_vct = profession.value_counts()
sta_vct = state.value_counts()
rf_vct = risk_flag.value_counts()
```

## 2.2 Thống kê mô tả

```python
df.describe()
```

```python

plt.pie(ms_vct.values, labels=ms_vct.index, autopct='%1.1f%%', startangle=90)
plt.title('Phân bố Married/Single')
plt.axis('equal')
plt.show()

plt.pie(ho_vct.values, labels=ho_vct.index, autopct='%1.1f%%', startangle=90)
plt.title('Phân bố House Ownership')
plt.axis('equal')
plt.show()

plt.pie(co_vct.values, labels=co_vct.index, autopct='%1.1f%%', startangle=90)
plt.title('Phân bố Car Ownership')
plt.axis('equal')
plt.show()
```

# Bước 3: Phân tích dữ liệu

## 3.1 Phân tích biến liên tục: Income, Age, Experience

```python
sns.histplot(income, kde=True, bins=20)
plt.title('Income Distribution')
plt.show()

sns.histplot(age, kde=True, bins=20)
plt.title('Age Distribution')
plt.show()

sns.histplot(experience, kde=True, bins=20)
plt.title('Experience Distribution')
plt.show()
```

## 3.2 Phân tích biến rời rạc

### 3.2.1 Phân tích Married/Single

```python
sns.countplot(x=CN_MS, data=df, hue=CN_RF, order=ms_vct.index.tolist())
plt.title('Married/Single vs Risk Flag')
plt.show()
```

### 3.2.2 Phân tích House Ownership

```python
sns.countplot(x=CN_HO, data=df, hue=CN_RF, order=ho_vct.index.tolist())
plt.title('House Ownership vs Risk Flag')
plt.show()
```

### 3.2.3 Phân tích biến Car Ownership

```python
sns.countplot(x=CN_CO, data=df, hue=CN_RF, order=co_vct.index.tolist())
plt.title('Car Ownership vs Risk Flag')
plt.show()
```

### 3.2.4 Phân tích Profession

```python
plt.figure(figsize=(20, 13))
sns.countplot(y=CN_PRO, data=df, hue=CN_RF, order=pro_vct.index.tolist())
plt.title('Profession vs Risk Flag')
plt.show()
```

### 3.2.5 Phân tích State

```python
plt.figure(figsize=(20, 7))
sns.countplot(y=CN_ST, data=df, hue=CN_RF, order=sta_vct.index.tolist())
plt.title('State vs Risk Flag')
plt.show()
```
